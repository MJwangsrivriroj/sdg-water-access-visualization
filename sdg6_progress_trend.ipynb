# Project Vis 2: Water Access Progress Over Time

## Goal
This visual shows how access to safely managed drinking water (SDG 6.1.1) has changed over the years in selected countries.

## Data Source
- File: API_SH.H2O.SMDW.ZS_DS2_en_csv_v2_19500.csv
- Source: [World Bank â€“ SDG Indicator 6.1.1](https://data.worldbank.org/indicator/SH.H2O.SMDW.ZS)

## 1. Load and Filter the Dataset

- The same dataset from Project Vis 1 is used here, focusing on a few selected countries to show progress over time.
- These four countries were selected to compare different levels of access. The United States is a high-income country with consistently high access. Brazil and Bangladesh are developing countries with improving access. Nigeria represents a country that still faces challenges in water access.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset and skip metadata rows
df = pd.read_csv("API_SH.H2O.SMDW.ZS_DS2_en_csv_v2_19500.csv", skiprows=4)

# Choose countries to compare
countries = ['United States', 'Brazil', 'Nigeria', 'Bangladesh']

# Filter data for selected countries
df_filtered = df[df['Country Name'].isin(countries)]

# Keep only year columns from 2000 to 2022
year_columns = [str(y) for y in range(2000, 2023)]
df_trend = df_filtered[['Country Name'] + year_columns].set_index('Country Name').T
```

## 2. Visual: Line Chart of Water Access Over Time

```python
# Plot the line chart
plt.figure(figsize=(10, 6))
for country in df_trend.columns:
    plt.plot(df_trend.index, df_trend[country], marker='o', label=country)

plt.title('Progress of Safely Managed Drinking Water Access (2000â€“2022)')
plt.xlabel('Year')
plt.ylabel('Access (%)')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 3. Explanation

This line chart shows how water access has changed over time access from 2000 to 2022. The United States stays near 100%. Brazil shows steady improvement. Bangladesh also improved but more slowly and Nigeria starts much lower but the trend shows progress over time..
